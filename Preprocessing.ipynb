Fill With Average Value

```python
#Origin	Solids	Chloramines	Sulfate	Conductivity	Organic_carbon	Trihalomethanes	Turbidity	Hardness	ph	Survivability
from decimal import Decimal, getcontext
import pandas as pd
getcontext().prec = 10000
df=pd.read_csv("water_survive.csv")
valid_ph = df['ph'].dropna()
print(Decimal(pow(10,-valid_ph).sum()))
avg_ph = -np.log(pow(10,-valid_ph).sum() / (len(valid_ph)))
for i in df.Origin:
    if i=='river':
        df.replace(i,int(0),inplace=True)
    elif i=='lake':
        df.replace(i,int(1),inplace=True)
    elif i=='ocean':
        df.replace(i,int(1),inplace=True)
print(avg_ph)
# 첫 번째 루프: 'Origin' 컬럼 값 변경
df['ph'].fillna(avg_ph,inplace=True)
for col in df.columns:
    valid=df[col].dropna()
    if col !='ph':
        df[col].fillna(valid.mean(),inplace=True)
    if col=='Origin':
        df[col].fillna(int(valid.mean()),inplace=True)
#Origin	Solids	Chloramines	Sulfate	Conductivity	Organic_carbon	Trihalomethanes	Turbidity	Hardness	ph	Survivability
df.to_csv("avg.csv", index=False)
```

Filled According to ph

```python
from decimal import Decimal, getcontext
import pandas as pd
import numpy as np
getcontext().prec=10000
df=pd.read_csv("fixed.csv")
df=df.drop('Unnamed: 0',axis=1)
df.replace(['', ' '], np.nan, inplace=True)
df
```

```python
df['ph'] = df['ph'].round(1)
df
```

```python
idx=df.columns
aciditygroup=df.dropna(subset='ph')
aciditygroup=df.groupby('ph').mean()
aciditygroup.Origin=np.int64(round(aciditygroup.Origin,0))
aciditygroup.rename_axis(None,inplace=True)
aciditygroup.fillna(method='ffill',inplace=True)
```

```python
cols = list(df.columns)
df = df.dropna(subset=['ph'])
df= df.loc[~df.index.duplicated()]
cols.remove('ph')
for idx, row in df.iterrows():
    ph_value = row['ph']
    for col in cols:
        if pd.isna(row[col]):
            # Fill missing value with corresponding value from aciditygroup
            df.at[idx, col] = aciditygroup.loc[ph_value, col]
df.isnull().sum()
df.to_csv("ph_group_improved.csv")
```

Using Solids

```python
from decimal import Decimal, getcontext
import pandas as pd
import numpy as np
getcontext().prec=10000
df=pd.read_csv("fixed.csv")
df=df.drop('Unnamed: 0',axis=1)
df.replace(['', ' '], np.nan, inplace=True)
df
df['Solids'] = df['Solids'].round(1)
df
min_val, max_val = 0,2
idx=df.columns
aciditygroup=df.dropna(subset='Solids')
aciditygroup=df.groupby('Solids').mean()
aciditygroup.Origin=np.int64(round(aciditygroup.Origin,0))
aciditygroup.rename_axis(None,inplace=True)
aciditygroup.fillna(method='ffill',inplace=True)
cols = list(df.columns)
df = df.dropna(subset=['Solids'])
df= df.loc[~df.index.duplicated()]
cols.remove('Solids')
aciditygroup['Origin'] = aciditygroup['Origin'].clip(min_val, max_val)
for idx, row in df.iterrows():
    ph_value = row['Solids']
    for col in cols:
        if pd.isna(row[col]):
            # Fill missing value with corresponding value from aciditygroup
            df.at[idx, col] = aciditygroup.loc[ph_value, col]
df.isnull().sum()
df.to_csv("solids_group_improved.csv")
```
